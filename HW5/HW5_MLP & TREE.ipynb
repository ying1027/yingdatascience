```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn import tree
from sklearn.metrics import accuracy_score
import pandas as pd
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
upload=pd.read_csv('upload.csv')
train.head()
print(train.keys())
print(train.shape)
```

```python
#data information
print(train.info())
print(test.info())
```

ONE HOT
=====

```python
# One-Hot encoding 
list_cat = ['Geography', 'Gender']
train_onehot = pd.get_dummies(train, columns = list_cat, prefix = list_cat)
test_onehot=pd.get_dummies(test,columns=list_cat,prefix=list_cat)
display(test_onehot.head())
display(train_onehot.head())
```

刪掉無關值
=====

```python
#刪掉無關值
train_set = train_onehot.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
test_set=test_onehot.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
display(train_set.head())
display(test_set.head())
```

```python
features=list(train_set.drop(columns=['Exited']))
target='Exited'
display(features)
```

標準化
=========

```python
sc=StandardScaler()

#標準化
train_set[features]=sc.fit_transform(train_set[features])
display(train_set)

#標準化
test_set[features]=sc.transform(test_set[features])
display(test_set)
```

MLP
=====

```python
#MLP
mlp=MLPClassifier()
mlp_=mlp.fit(train_set[features],train_set[target])
predict_mlp=mlp_.predict(test_set[features])
predict_mlp
```

```python
#upload
upload_mlp=upload
upload_mlp['Exited']=predict_mlp
upload_mlp.to_csv('upload__mlp2.csv')
```

Decision tree
=======

```python
#decision_tree
clf = tree.DecisionTreeRegressor()
clf = clf.fit(train_set[features],train_set[target])
tree=clf.predict(test_set[features])
```

```python
#upload
upload_tree=upload
upload_tree['Excited']=tree
upload_tree.to_csv('upload_tree.csv')
```
