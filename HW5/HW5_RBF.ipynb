```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn import svm
import pandas as pd
from sklearn.metrics import accuracy_score
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
upload=pd.read_csv('upload.csv')
train.head()
print(train.keys())
print(train.shape)
```

One hot
=====

```python
# One-Hot encoding 
list_cat = ['Geography', 'Gender']
train_onehot = pd.get_dummies(train, columns = list_cat, prefix = list_cat)
test_onehot=pd.get_dummies(test,columns=list_cat,prefix=list_cat)
display(test_onehot.head())
display(train_onehot.head())
```

刪掉無關的編號值
=====

```python
train_set = train_onehot.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
test_set=test_onehot.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
display(train_set.head())
display(test_set.head())
```

```python
features=list(train_set.drop(columns=['Exited']))
target='Exited'
display(features)
```

標準化
========

```python
sc=StandardScaler()

#標準化
train_set[features]=sc.fit_transform(train_set[features])
display(train_set)

#標準化
test_set[features]=sc.transform(test_set[features])
display(test_set)
```

RBF
=======

```python
#rbf
mysvm=svm.SVC(kernel='rbf',gamma=0.04,C=0.85)
mysvm.fit(train_set[features],train_set[target])
predict_rbf=mysvm.predict(test_set[features])
display(predict_rbf)
```

```python
upload_rbf=upload
upload_rbf['Excited']=predict_rbf
upload_rbf.to_csv('uploadrbf.csv')
```

POLY
=======

```python
#POLY
mysvm1=svm.SVC(kernel='poly',gamma=0.04,C=0.85)
mysvm1.fit(train_set[features],train_set[target])
predict_poly=mysvm1.predict(test_set[features])
display(predict_poly)
```

```python
upload_poly=upload
upload_poly['Exited']=predict_poly
upload_poly.to_csv('uploadpoly.csv')
```

SIGMOID
===========

```python
mysvm2=svm.SVC(kernel='sigmoid',gamma=0.05,C=0.87)
mysvm2.fit(train_set[features],train_set[target])
predict_sigmoid=mysvm2.predict(test_set[features])
display(predict_sigmoid)
```

```python
upload_sigmoid=upload
upload_sigmoid['Exited']=predict_poly
upload_sigmoid.to_csv('uploadsigmoid.csv')
```
